```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
np.random.seed(seed=42)

from collections import Counter
```

Importing CSV files

```python
shops = pd.read_csv('C:\\Users\\Rose\\Desktop\\Capstone3\\shops.csv')
salestrain = pd.read_csv('C:\\Users\\Rose\\Desktop\\Capstone3\\sales_train.csv')
items = pd.read_csv('C:\\Users\\Rose\\Desktop\\Capstone3\\items.csv')
itemscat = pd.read_csv('C:\\Users\\Rose\\Desktop\\Capstone3\\item_categories.csv')
```

Joining item_name and item_category_name using item_category_id

```python
itemsandcat = pd.merge(itemscat, items, on='item_category_id')
itemsandcat
```

Joining item_name and item_category_name using item_category_id

```python
itemscatprice = pd.merge(itemsandcat, salestrain , on='item_id')
itemscatprice
```

```python
ps4consoles = itemscatprice.loc[itemscatprice.item_category_name == 'Игровые консоли - PS4']
ps4consoles
```

```python
xboxoneconsoles = itemscatprice.loc[itemscatprice.item_category_name == 'Игровые консоли - XBOX ONE']
xboxoneconsoles
```

T-Test

```python
# for item price
stats.ttest_ind(ps4consoles['item_price'], xboxoneconsoles['item_price'], equal_var = False)
```

```python
# for item cnt day
stats.ttest_ind(ps4consoles['item_cnt_day'], xboxoneconsoles['item_cnt_day'], equal_var = False)
```

```python
#copying the merged CSV's dateframe to filter for year
ogdf = itemscatprice.copy()
ogdf
```

```python
#filtering date column for 2014 (couldn't figure out to do it by year so looked up if there was a ilike thing for pandas/python)
#assinging the query to a variable
df2014 = ogdf.query('date.str.contains("2014")', engine='python')
df2014
```

Masks

```python
ps4c = df2014.loc[df2014.item_category_name == 'Игровые консоли - PS4']
ps4c
```

```python
xboxonec = df2014.loc[df2014.item_category_name == 'Игровые консоли - XBOX ONE']
xboxonec
```

T-test

```python
# for item price
stats.ttest_ind(ps4c['item_price'], xboxonec['item_price'], equal_var = False)
```

```python
# for item cnt day
stats.ttest_ind(ps4c['item_cnt_day'], xboxonec['item_cnt_day'], equal_var = False)
```

```python
pvalue=0.000630169289320595
# interpret
alpha = 0.05
if pvalue > alpha:
	print('PS4 sold more than Xbox One)')
else:
	print('Xbox one sold more than PS4')
```

```python
pvalue=0.000630169289320595
# interpret
alpha = 0.05
if pvalue > alpha:
	print('Xbox One had more item count than PS4) (fail to reject H0)')
else:
	print('PS4 had more item count than Xbox (reject H0)')
```

```python
pvalue=1.3918252161198493e-285
# interpret
alpha = 0.05
if pvalue > alpha:
	print('Sample looks Gaussian (fail to reject H0)')
else:
	print('Sample does not look Gaussian (reject H0)')
```
