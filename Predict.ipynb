T-Test

```python
# for item price
stats.ttest_ind(ps3consoles['item_price'], ps4consoles['item_price'], equal_var = False)
```

```python
# for item cnt day
stats.ttest_ind(ps3consoles['item_cnt_day'], ps4consoles['item_cnt_day'], equal_var = False)
```

```python
pvalue=1.661730755930333e-60
# interpret
alpha = 0.05
if pvalue > alpha:
	print('Sample looks Gaussian (fail to reject H0)')
else:
	print('Sample does not look Gaussian (reject H0)')
```
